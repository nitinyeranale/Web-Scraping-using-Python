```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
book_url = "https://books.toscrape.com/catalogue/page-{}.html" 
```

```python
HEADERS={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/132.0.0.0 Safari/537.36"}
```

```python
book_data=[]
```

```python
for page in range(1,5):
    url=book_url.format(page)
    webpage=requests.get(url,headers=HEADERS)
    print(f"Scraped page {page} successfully!")
```

```python
webpage
```

```python
type(webpage.content)
```

```python
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
soup
```

```python
books = soup.find_all("article", attrs={"class": "product_pod"})
```

```python
books
```

```python
# for book in books:
#     title=book.h3.a["title"]
#     # price=book.find("p",attrs={"class":"price_color"}).text.strip()  # For Multiple attributes
#     # stock=book.find("p",attrs={"class":"instock availability"}).text.strip()
#     book_data.append({"Title":title,"Price":price,"Stock":stock})

for book in books:
        title = book.h3.a["title"]
        price = book.find("p", class_="price_color").text.strip()
        stock = book.find("p", class_="instock availability").text.strip()

        # Store data in list
        book_data.append({"Title": title, "Price": price, "Stock": stock})
```

```python
df=pd.DataFrame(book_data)
```

```python
df.to_csv("book_data.csv",index=False)
```
